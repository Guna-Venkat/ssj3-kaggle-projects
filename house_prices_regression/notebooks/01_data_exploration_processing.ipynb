### 🔥 Handling `FireplaceQu`

````python
# Define mapping
fireplace_quality_map = {
    'NO': 0,  # No fireplace
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

# Fill missing values with 'NA' first
train_df['FireplaceQu'] = train_df['FireplaceQu'].fillna('NO')

# Map values to integers
train_df['FireplaceQu'] = train_df['FireplaceQu'].map(fireplace_quality_map)
````

### 📏 Handling `LotFrontage`
---

**`LotFrontage`**: *Linear feet of street connected to the property*

---

#### ⚠️ Can we assume missing `LotFrontage` as 0?

**No — not advisable.**

- A value of **0** implies *no street access*, which is **unlikely** for most properties.
- Most missing values are due to **data collection gaps**, not actual zero frontage.
- Imputing 0 would introduce **incorrect signals** to the model (like poor access).

---

#### ✅ Recommended Imputation Strategy

**Use the median `LotFrontage` per `Neighborhood`**  
This accounts for local variation in lot dimensions.

````python
# Group by neighborhood and fill missing LotFrontage with the median of that group
train_df['LotFrontage'] = train_df.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
````

### 🚗 Handling Garage-Related Features
---

The dataset contains several features related to garages. Many of these contain `NaN` values. However, **these are not missing values in the typical sense** — they indicate that a property **has no garage**.

---

#### 📋 List of Garage Features and NA Meaning

| Feature        | Type         | Description                                     | NA Meaning           |
|----------------|--------------|-------------------------------------------------|----------------------|
| `GarageType`   | Categorical  | Location of the garage                          | No Garage            |
| `GarageYrBlt`  | Numerical    | Year garage was built                           | No Garage            |
| `GarageFinish` | Categorical  | Interior finish of the garage                   | No Garage            |
| `GarageCars`   | Numerical    | Size of garage in car capacity                  | No Garage = 0        |
| `GarageArea`   | Numerical    | Garage area in square feet                      | No Garage = 0        |
| `GarageQual`   | Ordinal Cat  | Quality of the garage                           | No Garage            |
| `GarageCond`   | Ordinal Cat  | Condition of the garage                         | No Garage            |

---

### 🤔 Should We Use Imputers Like KNN?

**Not recommended.**

- KNN or other statistical imputers treat missing data as *unknown*.
- Here, `NaN` clearly means **"no garage"**, and imputing values (like year or area) would be misleading.
- Imputing garage year for a house without a garage is logically incorrect.

---

### ✅ Recommended Strategy: Contextual Imputation

Use **domain-aware imputation**:

#### 🔹 1. Categorical Garage Columns  
Replace `NaN` with `'None'` to indicate absence of a garage:

#### 🔹 2. Numerical Garage Columns
Replace NaN with 0 to indicate zero capacity/area/year built:

#### 🔢 Optional: Ordinal Encoding for Quality and Condition
For modeling purposes, encode garage quality/condition as ordinal integers


````python
# Categorical columns (fill NA with 'None')
garage_cat_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
train_df[garage_cat_cols] = train_df[garage_cat_cols].fillna('No')

# Numerical columns (fill NA with 0)
garage_num_cols = ['GarageYrBlt', 'GarageCars', 'GarageArea']
train_df[garage_num_cols] = train_df[garage_num_cols].fillna(0)

garage_qual_map = {
    'No': 0,
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

train_df['GarageQual'] = train_df['GarageQual'].map(garage_qual_map)
train_df['GarageCond'] = train_df['GarageCond'].map(garage_qual_map)

garage_finish_map = {
    'No': 0,
    'Unf': 1,
    'RFn': 2,
    'Fin': 3
}

train_df['GarageFinish'] = train_df['GarageFinish'].map(garage_finish_map)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def compute_vif(df, drop_cols=None, top_n=None, title="VIF Plot", sort=True):
    """
    Compute and plot VIF values.

    Parameters:
    - df: DataFrame (only numerical columns will be considered)
    - drop_cols: list of columns to drop before calculating VIF
    - top_n: number of top VIF features to show
    - title: title of the plot
    - sort: whether to sort by VIF descending
    """
    X = df.select_dtypes(include='number').copy()

    if drop_cols:
        X.drop(columns=drop_cols, inplace=True, errors='ignore')

    # Add constant
    X_const = add_constant(X)

    # Replace inf/nan
    X_const.replace([np.inf, -np.inf], np.nan, inplace=True)
    X_const.dropna(inplace=True)

    # VIF calculation
    vif_data = pd.DataFrame()
    vif_data['Feature'] = X_const.columns
    vif_data['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

    # Remove constant from plot
    vif_data = vif_data[vif_data['Feature'] != 'const']

    if sort:
        vif_data = vif_data.sort_values(by='VIF', ascending=False)

    if top_n:
        vif_data = vif_data.head(top_n)

    # Plot
    plt.figure(figsize=(30,10))
    plt.barh(vif_data['Feature'], vif_data['VIF'], color='skyblue')
    plt.xlabel('VIF Value')
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.grid(axis='x')
    plt.tight_layout()
    plt.show()

    return vif_data
````

### 🔍 Multicollinearity Detection & Feature Pruning Using VIF
---

#### 📌 Goal:
To reduce multicollinearity in the dataset using **Variance Inflation Factor (VIF)**.  
High VIF indicates strong linear relationships between predictors, which can hurt model performance, especially for linear models.

---

### 🧮 Step 1: VIF Analysis

Using `statsmodels.variance_inflation_factor`, we computed VIF for all numerical features.

#### ⚠️ Result:
Several features showed **`inf` (infinite)** VIF, indicating **perfect multicollinearity**:

| Feature        | VIF     |
|----------------|---------|
| BsmtFinSF1     | ∞       |
| BsmtFinSF2     | ∞       |
| BsmtUnfSF      | ∞       |
| TotalBsmtSF    | ∞       |
| 1stFlrSF       | ∞       |
| 2ndFlrSF       | ∞       |
| LowQualFinSF   | ∞       |
| GrLivArea      | ∞       |

---

### 🔗 Step 2: Multicollinearity Cause

These features are **mathematical composites**:

- `TotalBsmtSF = BsmtFinSF1 + BsmtFinSF2 + BsmtUnfSF`
- `GrLivArea = 1stFlrSF + 2ndFlrSF + LowQualFinSF`

Including both the **components and their totals** introduces **perfect linear dependence** — causing `inf` VIF.

---

### ✂️ Step 3: Feature Pruning Strategy

To resolve this, we **retain the total/composite** (more interpretable and compact), and drop the components:

````python
features_to_drop = [
    'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF',
    '1stFlrSF', '2ndFlrSF', 'LowQualFinSF'
]

train_df.drop(columns=features_to_drop, inplace=True)
````
